We will be implementing three algorithms - 

> 1. Naive/ Brute force algorithm
> 2. Knuth Morris Pratt algorithm
> 3. Boyer Morse Horspool algorithm

```python
import re
import time 
```

```python
class algo:
    
    def __init__(self,genome):                  # constructor saves genome as instance variable

        self.genome = genome 
    
    def brute_force(self, text, target):
        result = [] 
        m = len(text)
        n = len(target)
        i, j = 0, 0
        
        for i in range(0, m-n+1):
            for j in range(0, n):
                if text[i+j] != target[j]:
                    break
            
                if j == n-1:
                    result.append(i)
                
        if result:
            print("Pattern:", target)
            print("Sequence found at position index:- ",result)
        else:
            print("Sequence not found.")
            
            
    def π_table(self, pattern):
        
        n = len(pattern)
        lps = [0]
        
        for i in range(1, n):
            j = lps[i-1]
            
            while j>0 and pattern[j] != pattern[i]:
                j = lps[j-1]
                
            if pattern[j] == pattern[i]:
                lps.append(j+1)
            else:
                lps.append(j)
            
        return lps
         
        
    def knuth_morris_pratt(self, text, pattern):
    
        lps = self.π_table(pattern)                 
        result, j = [], 0
        m, n = len(text), len(pattern)
    
        for i in range(m):
            while j>0 and text[i] != pattern[j]:
                j = lps[j-1]
                
            if text[i] == pattern[j]: 
                j += 1
                
            if j == n: 
                result.append((i-j)+1)
                j = 0  
                
        if result:
            print("Pattern:", pattern)
            print("LPS:", lps)
            print("Sequence found at position index:- ", result)
        else:
            print("Sequence not found.")
        
            return result

            
    def boyer_morse_horspool(self,target):

        results=[] 
        forward_skipper={c:len(target)-i-1 for i,c in enumerate(target[:-1])}
        backward_skipper={c:len(target)-i-1 for i,c in enumerate(target[:0:-1])}
        
        if(sum(forward_skipper.values())<sum(backward_skipper.values())):         # compare forward and backwards BMH

            current_index=len(target)-1                                           # forward BMH                
            while(current_index<len(self.genome)):      

                flag=1                                          
                for i in range(len(target)):                          
                    genome_character=self.genome[current_index-i]      
                    target_character=target[len(target)-i-1]    
                    if genome_character!=target_character:    
                        flag=0                                     
                        break

                if flag==1:                                         
                    results.append(current_index-i)

                skip=forward_skipper.get(self.genome[current_index],len(target))  # backwards BMH
                current_index+=skip                                 

        else:
            
            current_index=len(self.genome)-len(target)  
            while(current_index>0):      

                flag=1                                          
                for i in range(len(target)):                          
                    genome_character=self.genome[current_index+i]      
                    target_character=target[i]    
                    if genome_character!=target_character:    
                        flag=0                                      
                        break

                if flag==1:                                         
                    results.append(current_index)

                skip=backward_skipper.get(self.genome[current_index],len(target)) 
                current_index-=skip                                      
        
        results.sort(reverse=False)
        print("Pattern:",target)
        print("Skipper:",forward_skipper,backward_skipper)
        print("Sequence found at position index:- ",end=' ')
        print(results,sep=', ')

```

```python
if __name__ == '__main__':
    
    search = algo(dna)
    
    print("--------------------------------")
    print("Brute Force Algorithm")
    print("--------------------------------")
    print("")
    
    start = time.time()
    search.brute_force(dna, pattern)
    end = time.time()
    
    naive_timetaken = "%.8f"%(end-start)
    
    print("")
    print("Time elapsed = ", naive_timetaken, "seconds")
    print("")
    
    print("--------------------------------")
    print("Knuth Morris Pratt Algorithm")
    print("--------------------------------")
    print("")
    
    start = time.time()
    search.knuth_morris_pratt(dna, pattern)
    end = time.time()
    
    kmp_timetaken = "%.8f"%(end-start)
    
    print("")
    print("Time elapsed = ", kmp_timetaken, "seconds")
    print("")
    
    print("--------------------------------")
    print("Boyer Morse Horspool Algorithm")
    print("--------------------------------")
    print("")
    
    start = time.time()
    search.boyer_morse_horspool(pattern)
    end = time.time()
    
    bmh_timetaken = "%.8f"%(end-start)
    
    print("")
    print("Time elapsed = ", bmh_timetaken, "seconds")
    print("")
```
